### Organizador de Arquivos Automatizado

Este programa Python irá organizar arquivos em pastas com base em suas extensões. Ele irá:
1.  Definir diretórios de origem e destino.
2.  Criar arquivos de teste (10 de cada tipo) para demonstrar a funcionalidade.
3.  Varrer o diretório de origem.
4.  Criar pastas de destino com base nas extensões dos arquivos.
5.  Mover os arquivos para suas respectivas pastas.

**Atenção:** Este programa move arquivos. Recomenda-se testá-lo com arquivos de exemplo ou em um diretório não essencial primeiro.

```python
import os
import shutil
import random

# Exibir uma mensagem informando que os módulos foram importados.
print("Módulos 'os', 'shutil' e 'random' importados com sucesso para manipulação de arquivos e geração de dados de teste.")
```

```python
# --- Configuração dos Diretórios ---

# Define o diretório de origem onde os arquivos desorganizados estarão.
# Usamos '/content/source_files' como padrão no Colab.
source_directory = '/content/source_files'

# Define o diretório de destino onde as pastas organizadas serão criadas.
# Usamos '/content/organized_files' como padrão no Colab.
destination_directory = '/content/organized_files'

# Lista de extensões de arquivo para criar arquivos de teste e organizar.
# O usuário pode expandir esta lista conforme suas necessidades.
file_extensions = ['.pdf', '.jpg', '.png', '.doc', '.docx', '.xls', '.xlsx', '.py', '.txt', '.zip', '.mp3', '.mp4']

# Remove os diretórios existentes (e seu conteúdo) para garantir um teste limpo cada vez que o código for executado.
# Isso é útil para testes, mas **cuidado** ao usar em ambientes de produção.
if os.path.exists(source_directory):
    shutil.rmtree(source_directory)
    print(f"Diretório de origem existente '{source_directory}' removido para reinício limpo.")
if os.path.exists(destination_directory):
    shutil.rmtree(destination_directory)
    print(f"Diretório de destino existente '{destination_directory}' removido para reinício limpo.")

# Cria os diretórios de origem e destino vazios.
os.makedirs(source_directory)
os.makedirs(destination_directory)

print(f"Diretório de origem criado: {source_directory}")
print(f"Diretório de destino criado: {destination_directory}")
```

```python
# --- Função para Criar Arquivos de Teste ---

def create_dummy_files(num_files_per_type=10):
    print(f"\nCriando {num_files_per_type} arquivos de teste para cada extensão em '{source_directory}'...")
    for ext in file_extensions:
        # Cria 10 arquivos de cada tipo com conteúdo simples
        for i in range(1, num_files_per_type + 1):
            filename = f"documento_teste_{i}{ext}"
            filepath = os.path.join(source_directory, filename)
            # Cria um arquivo com um conteúdo de texto básico para simular um arquivo real.
            with open(filepath, 'w') as f:
                f.write(f"Este é um arquivo de teste {ext} número {i}.\n")
            print(f"  Criado: {filename}")
    print("Criação de arquivos de teste concluída.")

# Executa a função para popular o diretório de origem com arquivos de teste.
create_dummy_files(num_files_per_type=10)
```

```python
# --- Função Principal de Organização de Arquivos ---

def organize_files_by_extension(source_dir, dest_dir):
    # Verifica se o diretório de origem existe. Se não, imprime um erro e sai.
    if not os.path.isdir(source_dir):
        print(f"Erro: O diretório de origem '{source_dir}' não existe.")
        return

    print(f"\nIniciando a organização de arquivos do diretório: '{source_dir}' para '{dest_dir}'...")

    # Itera sobre todos os itens (arquivos e pastas) no diretório de origem.
    for filename in os.listdir(source_dir):
        # Constrói o caminho completo para o item atual no diretório de origem.
        source_path = os.path.join(source_dir, filename)

        # Verifica se o item é um arquivo (e não uma pasta).
        if os.path.isfile(source_path):
            # Obtém o nome do arquivo e sua extensão.
            # os.path.splitext('arquivo.pdf') retorna ('arquivo', '.pdf')
            _, extension = os.path.splitext(filename)

            # Converte a extensão para minúsculas e remove o ponto inicial (ex: '.pdf' -> 'pdf').
            # Se não houver extensão (arquivo sem ponto), usa 'sem_extensao' como nome da pasta.
            folder_name = extension[1:].lower() if extension else 'sem_extensao'

            # Constrói o caminho completo para a nova pasta de destino para este tipo de arquivo.
            destination_folder_path = os.path.join(dest_dir, folder_name)

            # Verifica se a pasta de destino para esta extensão já existe.
            # Se não existir, a cria.
            if not os.path.exists(destination_folder_path):
                os.makedirs(destination_folder_path)
                print(f"  Pasta '{folder_name}' criada em '{dest_dir}'.")

            # Constrói o caminho completo para onde o arquivo será movido.
            destination_path = os.path.join(destination_folder_path, filename)

            # Tenta mover o arquivo. Se ocorrer um erro (ex: permissão negada), ele será capturado.
            try:
                shutil.move(source_path, destination_path)
                print(f"  Movido: '{filename}' para '{folder_name}/'.")
            except Exception as e:
                print(f"  Erro ao mover '{filename}': {e}")
        else:
            # Ignora diretórios e outros itens que não são arquivos.
            print(f"  Ignorando item que não é arquivo ou é diretório: '{filename}'.")

    print("Organização de arquivos concluída com sucesso!")

# Executa a função principal para organizar os arquivos.
organize_files_by_extension(source_directory, destination_directory)
```

Desenvolver um script que consulte uma API pública (correios) e retorne as informações de endereço formatada em JSON. Use tratamento de exceção

```python
import requests
import json

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço de um determinado CEP.
    Retorna um dicionário com os dados do endereço ou informações de erro.
    """
    # Limpa o CEP, removendo caracteres não numéricos
    cep = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP tem o formato correto (8 dígitos)
    if len(cep) != 8:
        return {"erro": True, "mensagem": "CEP inválido. Deve conter 8 dígitos numéricos."}

    # Constrói a URL da API ViaCEP
    url = f"https://viacep.com.br/ws/{cep}/json/"

    print(f"Consultando CEP: {cep}...")

    try:
        # Faz a requisição HTTP para a API
        response = requests.get(url, timeout=5) # Adiciona um timeout para evitar esperas infinitas
        response.raise_for_status() # Lança um erro para códigos de status HTTP ruins (4xx ou 5xx)

        # Decodifica a resposta JSON
        dados_endereco = response.json()

        # Verifica se a API retornou um erro (ex: CEP não encontrado)
        if 'erro' in dados_endereco and dados_endereco['erro'] == True:
            return {"erro": True, "mensagem": "CEP não encontrado ou inexistente."}
        else:
            # Retorna os dados do endereço formatados
            return {
                "erro": False,
                "mensagem": "Sucesso",
                "dados": {
                    "cep": dados_endereco.get('cep'),
                    "logradouro": dados_endereco.get('logradouro'),
                    "complemento": dados_endereco.get('complemento'),
                    "bairro": dados_endereco.get('bairro'),
                    "localidade": dados_endereco.get('localidade'),
                    "uf": dados_endereco.get('uf'),
                    "ibge": dados_endereco.get('ibge'),
                    "gia": dados_endereco.get('gia'),
                    "ddd": dados_endereco.get('ddd'),
                    "siafi": dados_endereco.get('siafi')
                }
            }

    except requests.exceptions.ConnectionError:
        return {"erro": True, "mensagem": "Erro de conexão: Verifique sua internet."}
    except requests.exceptions.Timeout:
        return {"erro": True, "mensagem": "Tempo limite excedido: A API demorou muito para responder."}
    except requests.exceptions.HTTPError as e:
        return {"erro": True, "mensagem": f"Erro HTTP: {e.response.status_code} - {e.response.reason}"}
    except requests.exceptions.RequestException as e:
        return {"erro": True, "mensagem": f"Erro na requisição: {e}"}
    except json.JSONDecodeError:
        return {"erro": True, "mensagem": "Erro ao decodificar a resposta JSON da API."}
    except Exception as e:
        return {"erro": True, "mensagem": f"Um erro inesperado ocorreu: {e}"}

# --- Exemplo de Uso ---
if __name__ == '__main__':
    # Solicita ao usuário um CEP
    cep_digitado = input("Por favor, digite um CEP para consulta (ex: 01001000): ")

    # Chama a função de consulta
    resultado = consultar_cep(cep_digitado)

    # Imprime o resultado formatado como JSON
    print("\n--- Resultado da Consulta ---")
    print(json.dumps(resultado, indent=4, ensure_ascii=False))

    # Exemplo com CEP inválido
    # print("\n--- Teste com CEP inválido ---")
    # resultado_invalido = consultar_cep("12345")
    # print(json.dumps(resultado_invalido, indent=4, ensure_ascii=False))

    # Exemplo com CEP não existente (normalmente retorna erro da API ou objeto vazio)
    # print("\n--- Teste com CEP não existente (ex: 99999999) ---")
    # resultado_nao_existente = consultar_cep("99999999")
    # print(json.dumps(resultado_nao_existente, indent=4, ensure_ascii=False))
```
